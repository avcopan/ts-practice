```python
import automol
from tblite.ase import TBLite
from ase.optimize import BFGS

geo = automol.geom.from_data(
    symbs=["O", "H", "H"],
    xyzs=[[0., 0., 0.], [1., 0., 0.], [0., 1., 0.]],
    angstrom=True
)
automol.geom.display(geo)

atms_obj = automol.geom.ase_atoms(geo)
atms_obj.calc = TBLite(method="GFN1-xTB")

opt_eq = BFGS(atms_obj)
opt_eq.run()

geo_out = automol.geom.from_ase_atoms(atms_obj)
automol.geom.display(geo_out)
```
